```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 gas_one = pd.read_csv('gas_prices.csv')
 print(gas_one)
```

```python
 plt.plot(gas_one.Year, gas_one.USA)
 plt.plot(gas_one.Year, gas_one.Canada)
 plt.plot(gas_one.Year, gas_one['South Korea'])
 plt.title('Gas prices over Time')
 plt.xlabel('Year')
 plt.ylabel('Price')
 plt.legend(['USA','Canada','South Korea'])
 print(gas_one.Year[::3])
 plt.xticks(gas_one.Year[::3])
 plt.figure(figsize=(10,5))

 plt.show()
```

```python
foot_ball = pd.read_csv('fifa_data.csv')
foot_ball
```

```python
plt.plot(foot_ball.Name, foot_ball.Age)

plt.title('Football Selection graph')
plt.xlabel('Name')
plt.ylabel('Age')
plt.legend(['Age','Overall'])
plt.figure(figsize=(15,15))

for football in foot_ball:
    plt.plot(foot_ball.Name, foot_ball[football], marker= '.')
    plt.xticks(foot_ball.Name[::3])
```

```python

age = [0,10,20,30,40,50,60,70,80]
plt.hist(foot_ball.Age ,bins=age)
plt.title('Football Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```

```python
foot_ball['Preferred Foot'].value_counts()
left_leg = foot_ball['Preferred Foot'].value_counts()['Left']
right_leg = foot_ball['Preferred Foot'].value_counts()['Right']
print(left_leg,right_leg)
```

```python
plt.pie([left_leg,right_leg], autopct='%.2f%%')
plt.legend(['Left','Right'])
plt.title('Preferred Foot')
plt.show()
plt.show()
```
